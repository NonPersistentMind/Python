```
import numpy as np

def mandelbrot_set(height, width, real_bot, real_top, imag_bot, imag_top, iterations=200, bound=5):
  real_vals = np.linspace(real_bot, real_top, width)
  imag_vals = np.linspace(imag_bot, imag_top, height)

  result = np.ones((height,width), dtype=np.float32)

  for i in range(height):

    for j in range(width):
      
      c = np.complex64(real_vals[j] + imag_vals[i] * 1j)
      z = np.complex64(0)

      for _ in range(iterations):
        z = z**2 + c 
        if (np.abs(z) > bound):
          result[i,j] = 0
          break
  
  # np.flip(result, axis=0)
  return result


      
```

```
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.use('Agg')

layout = (5,15,-10,0)

plt.figure(1, figsize=(10,10), dpi=80)
graph = mandelbrot_set(1000,1000,*layout, bound=30)
plt.imshow(graph, extent=layout)
```
