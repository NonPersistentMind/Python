```
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats as st
import math
plt.style.use('ggplot')

sqrt = math.sqrt
PI = math.pi
arr = np.array
seq = np.arange
N = st.norm()

def plot_fig(x,y, step=False):
  if step:
    print('Mean:', np.dot(x,y)*step)
  plt.plot(x,y)
  fig = plt.gcf()
  fig.set_size_inches(15,7.5)
  plt.show()

def test_hypothesis():
  print('Needed_Z:', -N.ppf(alpha/sides))
  print('Current_Z:', (sample_mean - mean)*sqrt(n)/std )
  return (sample_mean - mean)*sqrt(n)/std

def get_p_value():
  z_transform = (sample_mean - mean)/(std/sqrt(n)) 
  print('P value:', N.cdf(-abs(z_transform))*sides)
  return z_transform, N.cdf(-abs(z_transform))*sides


def large_sample_test():
  std = sqrt(n*p*(1-p))
  needed_z = abs(N.ppf(alpha/sides))
  current_z = abs((success - n*p) / std)
  return current_z, needed_z
```

# Power of Test

```
mean = 494
true_mean = 500
n = 86
std = 124/sqrt(n)
alpha = 0.05
sides = 2
pos_crit_val = mean + std*abs(N.ppf(alpha/sides)) 
neg_crit_val = mean - std*abs(N.ppf(alpha/sides)) 
print('Power: ',1 - N.cdf((pos_crit_val - true_mean)/std) + N.cdf((neg_crit_val - true_mean)/std))
```

```
mean = 95
true_mean = 90
n = 22
std = 15/sqrt(n)
alpha = 0.06
sides = 2

pos_crit_val = mean + std*abs(N.ppf(alpha/sides)) 
neg_crit_val = mean - std*abs(N.ppf(alpha/sides))

print(pos_crit_val, neg_crit_val)
print('Power:',  N.cdf((neg_crit_val - true_mean)/std) + 1 - N.cdf((pos_crit_val - true_mean)/std))
```

```
mean = 60
n = 16
std = 4/sqrt(n)
alpha = 0.05
sides = 2

pos_crit_val = mean + std*abs(N.ppf(alpha/sides)) 
neg_crit_val = mean - std*abs(N.ppf(alpha/sides))

x = np.arange(30, 90, 0.01)
y = [N.cdf((neg_crit_val - true_mean)/std) + 1 - N.cdf((pos_crit_val - true_mean)/std) for true_mean in x]

plot_fig(x,y)
```
