```
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats as st
import math
plt.style.use('ggplot')

sqrt = math.sqrt
PI = math.pi
arr = np.array
arange = np.arange

def conf_int(suc_num, sample_num, alpha, finite_correction=False, population_num=None):
  Y = st.norm()
  z = Y.ppf(1-alpha/2)
  sample_avg = suc_num/sample_num
  correction = 1
  if finite_correction:
    correction = (population_num - sample_num)/(population_num - 1)

  print('Correction: ', correction)

  std = sqrt(sample_avg*(1-sample_avg)*correction/sample_num)
  ar = arr([-z,z])
  return (sample_avg + ar*std)

def plot_fig(x,y, step=False):
  if step:
    print('Mean:', np.dot(x,y)*step)
  plt.plot(x,y)
  fig = plt.gcf()
  fig.set_size_inches(15,7.5)
  plt.show()
```

# Estimator's plotting


```
def unbiased_y_min(x, theta, sample_size):
  t = theta
  n = sample_size
  return n/(n+1) * 1/t * (1 - x/(t*(n+1)))**(n-1)

theta = 5
step = 0.01
n = 100
x = seq(0,theta*(n+1)+step, step)
y = unbiased_y_min(x, theta, n)


plot_fig(x,y, step)
```

```
def y_3(x, theta):
  t = theta
  return 36/5 * ((3/5*x)**2*t - (3/5*x)**3)/t**4

theta = 6
step = 0.01
x = seq(0, theta*(5/3)+step, step)
y = y_3(x, theta)
plot_fig(x,y, step)
```
