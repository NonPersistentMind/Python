```
import scipy.stats as st
import math
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

def norm(mean = 0, std = 1):
  Z = st.norm(mean, std)
  P = Z.cdf
  return Z, P

def sqrt(x):
  return math.sqrt(x)

PI = math.pi
```

```
R = st.expon(0.75)
C = st.norm(100,20)

samples = 50000
range_samples = list(R.rvs(samples))
ranges = []
ran = []
interval_len = 0

while len(range_samples):
  if range_samples[0] > 2:
    range_samples.pop(0)
  elif sum(ran) + range_samples[0] < 2:
    ran.append(range_samples.pop(0))
  else:
    ranges.append(ran)
    ran = []

prices = []
for ran in ranges:
  
  prices.append(sum(C.rvs(len(ran))))  
```

```
fig, ax = plt.subplots(1,1)
fig.set_size_inches(15,7.5)

plt.hist(prices, bins = np.arange(min(prices),max(prices), 2), edgecolor='g', density=True)
plt.show()
```
