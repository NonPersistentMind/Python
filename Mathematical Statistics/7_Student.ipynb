```
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats as st
import math
plt.style.use('ggplot')

sqrt = math.sqrt
PI = math.pi
arr = np.array
arange = np.arange

def conf_int(suc_num, sample_num, alpha, finite_correction=False, population_num=None):
  Y = st.norm()
  z = Y.ppf(1-alpha/2)
  sample_avg = suc_num/sample_num
  correction = 1
  if finite_correction:
    correction = (population_num - sample_num)/(population_num - 1)

  print('Correction: ', correction)

  std = sqrt(sample_avg*(1-sample_avg)*correction/sample_num)
  ar = arr([-z,z])
  return (sample_avg + ar*std)

def plot_fig(x,y, step=False):
  if step:
    print('Mean:', np.dot(x,y)*step)
  plt.plot(x,y)
  fig = plt.gcf()
  fig.set_size_inches(15,7.5)
  plt.show()

def plot_figs(x, *ys):
  for y_i in ys:
    plt.plot(x,y_i)
  fig = plt.gcf()
  fig.set_size_inches(15,7.5)
  plt.show()
```

```
x = arange(0,1,0.01)
y_1 = 2*x
y_2 = 4*x**3
plot_figs(x, y_1, y_2)
```

```
ar = arr([65,55,30])
mean = 50
std = 10
ar = (ar - mean)/std
print(ar)
s = sum(ar**2)
S = st.chi2(3)
print(s)
print(1 - S.cdf(s))
x = arange(0,20,0.01)
y = S.pdf(x)
plot_fig(x,y)
```
